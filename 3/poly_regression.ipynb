## Implement Polynomial Regression

```python
# import important libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Import the data
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
Y = dataset.iloc[:, 2].values
```

```python
#fitting linear regression model in dataset
from sklearn.linear_model import LinearRegression
linearreg=LinearRegression()
linearreg.fit(X,Y)
```

```python
#fitting polynomial regression model in dataset
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(X)
second_lr=LinearRegression()
second_lr.fit(X_poly,Y)
```

```python
#visualizing linear regression result
plt.scatter(X,Y,color='red')
plt.plot(X,linearreg.predict(X),color='blue')
plt.title('Truth or Bluff Linear Regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
#visualizing polynomial regression result
X_grid=np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,Y,color='red')
plt.plot(X_grid,second_lr.predict(poly_reg.fit_transform(X_grid)),color='blue')
plt.title('Truth or Bluff Polynomial Regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()

```
